```python
import time
import datetime
file = open('out3.txt','w') 
file.write('iteration,committed,accepted,not_accepted' + '\n')
for iter in d2:  
    pts_c = 0
    pts_a = 0
    pts_u = 0
    #print('iteration: ', iter['number'])
    for iter2 in iter['stories']:
        pts_c += iter2.get('estimate', 0)
        #print('\tstory-id: ', iter2['id'])
        if(iter2['current_state'] == 'accepted'):
            #end_date.split(" ")
            #print('\tfinish: ', iter['finish'].split("T")[0],'\n\taccepted: ', iter2.get('accepted_at', '2020-01-01T').split("T")[0], '\ttype: ')
            iFinish = time.strptime(iter['finish'].split("T")[0],"%Y-%m-%d")
            sFinish = time.strptime(iter2.get('accepted_at', '2020-01-01T').split("T")[0], "%Y-%m-%d")
           
            #print('\tiFinish: ', iter['finish'].split("T")[0])
            #print('\tsFinish: ', iter2.get('accepted_at', '2020-01-01T').split("T")[0])
            if(sFinish <= iFinish):
                file.write('id: ' + str(iter2['id']) + '\tcurrent-state: ' + iter2['current_state'] + '\tdate-check: ' + str(sFinish <= iFinish) + ' accepted during sprint++++' + '\n')
                pts_a += iter2.get('estimate', 0)
            else:
                z = 0
                file.write('id: ' + str(iter2['id']) + ' current-state: ' + iter2['current_state'] +'\t' +  ' not accepted during sprint!!!!!!!!!!!!!!!!!!!!!!!!!!!' + '\n')
        else:
            pts_u += iter2.get('estimate', 0)
            file.write('id: ' + str(iter2['id']) + ' current-state: ' + iter2['current_state'] + '\t'+ ' not accepted during sprint###########################' + '\n')
    file.write(str(iter['number'])+','+str(pts_c)+','+str(pts_a)+','+str(pts_u) + '\n')
file.write('done...')   
file.close()
```
